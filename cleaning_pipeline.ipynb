# Merge Dataset

```python
raw_dataset = (
    df_application_metadata
      .merge(df_demographics, left_on='customer_ref', right_on='cust_id', how='inner', suffixes=('', '_demographics'))
      .drop(columns=['cust_id'])
      .merge(df_loan_details, left_on='customer_ref', right_on='customer_id', how='inner', suffixes=('', '_loan_details'))
      .drop(columns=['customer_id'])
      .merge(df_credit_history, left_on='customer_ref', right_on='customer_number', how='inner', suffixes=('', '_credit_history'))
      .drop(columns=['customer_number'])
      .merge(df_financial_ratios, left_on='customer_ref', right_on='cust_num', how='inner', suffixes=('', '_financial_ratios'))
      .drop(columns=['cust_num'])
      .merge(df_geographic_data, left_on='customer_ref', right_on='id', how='inner', suffixes=('', '_geographic_data'))
      .drop(columns=['id'])
)
```

# Clean & Analyze

```python
import re, unicodedata
import numpy as np


NA_SET = {"", "na", "n/a", "none", "null", "-", "", "unknown", "?", "nan",
          " ", " ", "", "yo'q", "yoq", "yo`q", "yoq", "yoq"}
TRUE_SET  = {"true","t","yes","y","1","ok","","","","ha","xa","bor"}
FALSE_SET = {"false","f","no","n","0","","","yo'q","yoq","yo`q","yoq","yoq","yok",""}

def norm_text(v):
    if pd.isna(v): return np.nan
    s = unicodedata.normalize("NFKC", str(v))
    s = re.sub(r"[\u200B-\u200D\uFEFF]", "", s)
    s = re.sub(r"\s+", " ", s).strip()
    return np.nan if s.lower() in NA_SET else s

def to_bool(v):
    if pd.isna(v): return pd.NA
    s = str(v).strip().lower()
    if s in TRUE_SET:  return True
    if s in FALSE_SET: return False
    return pd.NA

def to_number(v):
    if pd.isna(v): return np.nan
    s = str(v).strip()
    if s.startswith("$"):
      s=s[1:]
    if "," in s:
        s = s.replace(",", "")
    try:
        x = float(s)
        return x
    except ValueError:
        return np.nan

def parse_date(v):
    if pd.isna(v): return pd.NaT
    s = str(v).strip()
    for dayfirst in (False, True):
        try:
            return pd.to_datetime(s, errors="raise", dayfirst=dayfirst)
        except Exception:
            pass
    return pd.NaT

def clean_email(v):
    if pd.isna(v): return pd.NA
    t = norm_text(v)
    return t.lower() if isinstance(t, str) else t

def clean_loan_type(v):
    if pd.isna(v): return pd.NA
    t = norm_text(v)
    if t in ["personal","Personal Loan","PERSONAL"]:
        t="Personal"
    elif t in ["CC","CreditCard","credit card"]:
        t="Credit Card"
    elif t in ["MORTGAGE","mortgage"]:
        t="Mortgage"
    return t if isinstance(t, str) else t

def clean_employment_type(v):
    if pd.isna(v): return pd.NA
    t = norm_text(v)
    if t in ['CONTRACT',  'Contractor']:
        t="Contract"
    elif t in ['FT', 'FULL_TIME', 'Full-time', 'Fulltime']:
        t="Full Time"
    elif t in ['PART_TIME', 'PT', 'Part Time', 'Part-time']:
        t="Part Time"
    elif t in ['SELF_EMPLOYED', 'Self Emp','Self-employed']:
        t="Self Employed"
    return t if isinstance(t, str) else t

def clean_phone(v):
    if pd.isna(v): return pd.NA
    t = unicodedata.normalize("NFKC", str(v))
    t = re.sub(r"[^+\d]", "", t)
    t = re.sub(r"^\++", "+", t)
    return t or pd.NA
```

```python
clean_map = {
    'customer_ref': to_number,
    'application_id': to_number,
    'application_hour': to_number,
    'application_day_of_week': to_number,
    'account_open_year': to_number,
    'preferred_contact': norm_text,
    'referral_code': norm_text,
    'account_status_code': norm_text,
    'num_login_sessions': to_number,
    'num_customer_service_calls': to_number,
    'has_mobile_app': clean_phone,
    'paperless_billing': to_number,
    'default': to_number,
    'age': to_number,
    'annual_income': to_number,
    'employment_length': to_number,
    'employment_type': norm_text,
    "employment_type":clean_employment_type,
    'education': norm_text,
    'marital_status': norm_text,
    'num_dependents': norm_text,
    'loan_type': norm_text,
    'loan_type': clean_loan_type,
    'loan_amount': to_number,
    'loan_term': norm_text,
    'interest_rate': to_number,
    'loan_purpose': norm_text,
    'loan_to_value_ratio': to_number,
    'origination_channel': norm_text,
    'loan_officer_id': to_number,
    'marketing_campaign': norm_text,
    'credit_score': to_number,
    'num_credit_accounts': to_number,
    'oldest_credit_line_age': to_number,
    'oldest_account_age_months': to_number,
    'total_credit_limit': to_number,
    'num_delinquencies_2yrs': to_number,
    'num_inquiries_6mo': to_number,
    'recent_inquiry_count': to_number,
    'num_public_records': to_number,
    'num_collections': to_number,
    'account_diversity_index': to_number,
    'monthly_income': to_number,
    'existing_monthly_debt': to_number,
    'monthly_payment': to_number,
    'debt_to_income_ratio': to_number,
    'debt_service_ratio': to_number,
    'payment_to_income_ratio': to_number,
    'credit_utilization': norm_text,
    'revolving_balance': to_number,
    'credit_usage_amount': to_number,
    'available_credit': norm_text,
    'total_monthly_debt_payment': to_number,
    'annual_debt_payment': to_number,
    'loan_to_annual_income': to_number,
    'total_debt_amount': to_number,
    'monthly_free_cash_flow': to_number,
    'state': norm_text,
    'regional_unemployment_rate': to_number,
    'regional_median_income': to_number,
    'regional_median_rent': to_number,
    'housing_price_index': to_number,
    'cost_of_living_index': to_number,
    'previous_zip_code': norm_text,
    "available_credit":to_number
}

for col, fn in clean_map.items():
    if col in raw_dataset.columns:
        raw_dataset[col] = raw_dataset[col].apply(lambda x: fn(x))
```
